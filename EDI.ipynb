```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
import numpy as np
import math
from tensorflow.keras.preprocessing import image
model = load_model(r'C:\Users\francis\Desktop\keras_model.h5')
body_cascade = cv2.CascadeClassifier(r"C:\Users\francis\Desktop\haarcascade_fullbody.xml")
```

## Real Time Detection

```python
def classify(cropped):
    data=[]
    img=cv2.imread(cropped)
    img = cv2.resize(img, (224,224))
    X = image.img_to_array(img)
    X = (X.astype(np.float32) / 255.0) 
    X = np.expand_dims(X, axis=0)
    preds=model.predict(X)
    data.append(preds[0])
    max_val=np.argmax(data)
    print(max_val)
    return max_val

def final_class(prediction):
    one="Jumping"
    two="Kicking"
    three="Running"
    four="Man with bat"
    five="Man with gun"
    six="Man with knife"
    
    #loop through predicted output, crosschecked with desired output 
    if prediction==0:
        return one
    elif prediction==1:
        return two
    elif prediction==2:
        return three  
    elif prediction==3:
        return four
    elif prediction==4:
        return five
    elif prediction==5:
        return six
    
    
cap = cv2.VideoCapture(0)

while True:
    # Get webcam images
    ret, frame = cap.read()
    
    grey_image = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    
    body = body_cascade.detectMultiScale(grey_image , 1.2 , 2)
    
    for (x , y, w , h) in body:
        
        cv2.rectangle(frame, (x , y) , (x+w , y + h) , (0 , 255 , 255) , 2)
        
        cropped = frame[x:y , x+w:y + h]
        
        prediction =classify(cropped)
        print(prediction)
    
        output=final_class(prediction)
        print(output)

        #cv2.rectangle(frame, (top_left_x,top_left_y), (bottom_right_x,bottom_right_y), (0,255,0), 3)
        cv2.putText(frame, output,(50,50), cv2.FONT_HERSHEY_COMPLEX, 2 ,(0,255,0), 2)
        cv2.imshow('Object Detector using SIFT', frame)
        if cv2.waitKey(1) == 13: #13 is the Enter Key
            break

cap.release()
cv2.destroyAllWindows()
```
